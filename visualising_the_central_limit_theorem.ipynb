```python
import seaborn as sns
import pandas as pd
from matplotlib import pyplot as plt

df = sns.load_dataset('taxis')
var = 'tip'

print('True mean of {} is {}.'.format(var, df[var].mean()))
print('Underlying distribution of values for {}:'.format(var))    

plt.hist(x = df[var], bins = 100)
plt.show()
```

Definitely not a normal distribution...

```python
var = 'tip'
n = 1000 # sample size
s = 2000 # number of samples drawn
est_list = []

for i in range(s):
    est_list.append(df[var].sample(n).mean())

print('Sampling distribution of the mean of {}:'.format(var))    
    
plt.hist(x = est_list, bins = 50)
plt.show()
```

... Wow! The means are normally distributed! And they're centered around the true mean!

```python
var_list = ['tip', 'fare']
n = 1000 # sample size
s = 2000 # number of samples drawn
est_list = []

for i in range(s):
    df_samp = df[var_list].sample(n)
    est_list.append(df_samp.iloc[:, 0].corr(df_samp.iloc[:, 1]))
    
print('Sampling distribution of the Pearson coefficient between {} and {}:'.format(var_list[0], var_list[1]))    
    
plt.hist(x = est_list, bins = 50)
plt.show()
```

```python
print('True Pearson coefficient between {} and {} is {}.'.format(var_list[0], var_list[1], df[var_list].iloc[:, 0].corr(df[var_list].iloc[:, 1])))
```
